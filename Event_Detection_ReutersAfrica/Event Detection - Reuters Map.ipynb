```python
import pandas as pd
import folium
```

```python
path = '/Users/jkuypers/Documents/NOVA_IMS/Term_2/Text_Mining/FinalProject_Version_Oliver_w:Coords/CentralAfricanRepublicScraper/ReutersAfrica_analysed.csv'
df = pd.read_csv(path)
```

```python
#creating Map
ReutersMap = folium.Map(location=[4.0383, 21.7587], tiles='Mapbox Bright', zoom_start=3)

#adding markers (i.e events)
for idx, coord in df.iterrows():
    folium.Marker(location=[coord['lat'], coord['long']], 
                  tooltip=coord['title']).add_to(ReutersMap)

#show map
ReutersMap
```
